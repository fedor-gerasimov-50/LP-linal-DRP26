```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['axes.grid'] = True

def plot_parametrization(t_min, t_max, x, y, num_points=600, title=None, labels=None, colors=None, alpha=0.35, size=12):
    t = np.linspace(t_min, t_max, num_points)
    x_funcs = x if isinstance(x, (list, tuple)) else [x]
    y_funcs = y if isinstance(y, (list, tuple)) else [y]

    if len(x_funcs) != len(y_funcs):
        raise ValueError('x and y must contain the same number of parametrizations.')

    if labels is None:
        labels = [None] * len(x_funcs)
    if colors is None:
        colors = [None] * len(x_funcs)

    fig, ax = plt.subplots(figsize=(6, 6))

    for x_func, y_func, label, color in zip(x_funcs, y_funcs, labels, colors):
        ax.scatter(x_func(t), y_func(t), alpha=alpha, s=size, label=label, color=color)

    ax.axhline(0, color='black', linewidth=0.8)
    ax.axvline(0, color='black', linewidth=0.8)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_aspect('equal')

    if title is not None:
        ax.set_title(title)

    if any(label is not None for label in labels):
        ax.legend()

    plt.show()
```

# 1.1.2 Parameterizations of level curves

We want parameterizations of the curves

1. $y^2 - x^2 = 1$
2. $\dfrac{x^2}{4} + \dfrac{y^2}{9} = 1$

For the hyperbola $y^2 - x^2 = 1$, use the identity
$\cosh^2 t - \sinh^2 t = 1$:

- upper branch: $x(t) = \sinh t$, $y(t) = \cosh t$
- lower branch: $x(t) = \sinh t$, $y(t) = -\cosh t$, where $t \in \mathbb{R}$

Indeed,
$y(t)^2 - x(t)^2 = \cosh^2 t - \sinh^2 t = 1$.

For the ellipse $\dfrac{x^2}{4} + \dfrac{y^2}{9} = 1$, use
$\cos^2 t + \sin^2 t = 1$:

- $x(t) = 2\cos t$, $y(t) = 3\sin t$, where $t \in [0, 2\pi]$

Then
$\dfrac{x(t)^2}{4} + \dfrac{y(t)^2}{9} = \cos^2 t + \sin^2 t = 1$.


```python
# Plot for y^2 - x^2 = 1
plot_parametrization(
    -2,
    2,
    x=(np.sinh, np.sinh),
    y=(np.cosh, lambda t: -np.cosh(t)),
    title=r'$y^2 - x^2 = 1$',
    labels=(
        r'$(x(t), y(t)) = (\sinh t, \cosh t)$',
        r'$(x(t), y(t)) = (\sinh t, -\cosh t)$'
    ),
    colors=('tab:blue', 'tab:orange')
)
```

```python
# Plot for x^2/4 + y^2/9 = 1
plot_parametrization(
    0,
    2 * np.pi,
    x=lambda t: 2 * np.cos(t),
    y=lambda t: 3 * np.sin(t),
    title=r'$\dfrac{x^2}{4} + \dfrac{y^2}{9} = 1$',
    labels=(r'$(x(t), y(t)) = (2\cos t, 3\sin t)$',),
    colors=('tab:green',)
)
```

```python
plot_parametrization(
    0,
    np.pi / 2,
    x=lambda t: np.cos(t)**2,
    y=lambda t: np.sin(t)**2,
    num_points=100,
    colors=('tab:purple',)
)
```
